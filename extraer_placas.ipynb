```python
import os 
from ultralytics import YOLO
import cv2
import numpy as np
base_path = "data\images"
model= YOLO("../Placas/runs/detect/yolo11_obb_custom/weights/best.pt")
placas_folder= "data/placas"
files = os.listdir(base_path)
for folder,_,files in os.walk(base_path):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            img_path = os.path.join(folder, file)
            img = cv2.imread(img_path)
            if img is not None:
                results = model.predict(img, conf=0.8)
                for result in results:
                    img_result = result.plot()
                    points = result.boxes.xyxy.cpu().numpy()
                    if points.size > 0:
                        x1, y1, x2, y2 = points[0]
                        img_placa = img[int(y1):int(y2), int(x1):int(x2)]
                        if os.path.exists(placas_folder):
                            cv2.imwrite(os.path.join(placas_folder, file), img_placa)
                        else:
                            os.makedirs(placas_folder)
                            cv2.imwrite(os.path.join(placas_folder, file), img_placa)
            else:
                print(f"Could not read image: {img_path}")  
            print(os.path.join(folder, file))
```
